### 1.ATM Simulation System
Design and develop a Python program that simulates the basic operations of an ATM. 

#### Mandatory Features:
- User authentication (PIN-based login)
- Balance enquiry
- Cash deposit
- Cash withdrawal
#### Optional Features:
- Transaction history
- Change PIN
- Daily withdrawal limit
- Exit option

```python
# ATM Simulation System using Core Python

# Initial data
pin = "2468"
balance = 50000
daily_limit = 10000
withdrawn_today = 0
transactions = []

# User authentication
def login():
    for i in range(3):
        entered_pin = input("Enter your ATM PIN: ")
        if entered_pin == pin:
            print("Login successful!\n")
            return True
        else:
            print("Incorrect PIN. Try again.")
    print("Too many failed attempts. Card blocked.")
    return False

# Balance enquiry
def check_balance():
    print("Your current balance is:", balance)

# Cash deposit
def deposit():
    global balance
    amount = int(input("Enter amount to deposit: "))
    if amount > 0:
        balance += amount
        transactions.append(f"Deposited ₹{amount}")
        print("Deposit successful.")
    else:
        print("Invalid amount.")

# Cash withdrawal
def withdraw():
    global balance, withdrawn_today
    amount = int(input("Enter amount to withdraw: "))

    if amount <= 0:
        print("Invalid amount.")
    elif amount > balance:
        print("Insufficient balance.")
    elif withdrawn_today + amount > daily_limit:
        print("Daily withdrawal limit exceeded.")
    else:
        balance -= amount
        withdrawn_today += amount
        transactions.append(f"Withdrew ₹{amount}")
        print("Please collect your cash.")

# Change PIN
def change_pin():
    global pin
    old_pin = input("Enter old PIN: ")
    if old_pin == pin:
        new_pin = input("Enter new PIN: ")
        pin = new_pin
        print("PIN changed successfully.")
    else:
        print("Incorrect old PIN.")

# Transaction history
def show_transactions():
    if transactions:
        print("Transaction History:")
        for t in transactions:
            print("-", t)
    else:
        print("No transactions yet.")

# Main program
if login():
    while True:
        print("\n--- ATM MENU ---")
        print("1. Balance Enquiry")
        print("2. Cash Deposit")
        print("3. Cash Withdrawal")
        print("4. Transaction History")
        print("5. Change PIN")
        print("6. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            check_balance()
        elif choice == "2":
            deposit()
        elif choice == "3":
            withdraw()
        elif choice == "4":
            show_transactions()
        elif choice == "5":
            change_pin()
        elif choice == "6":
            print("Thank you for using the ATM. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")
```
